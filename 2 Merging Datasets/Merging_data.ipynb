```python
import xarray as xr
import pandas as pd
import os
import pyarrow.parquet as pq
import pyarrow as pa
```

## Transparency file converted to parquet and merged

```python
#Define function to convert NetCDF to Parquet
def convert_nc_to_parquet(nc_file, parquet_file):
    ds = xr.open_dataset(nc_file)
    df = ds.to_dataframe().reset_index()
    table = pa.Table.from_pandas(df)
    pq.write_table(table, parquet_file)
    print(f"Converted {nc_file} to {parquet_file}")

#Convert each dataset to Parquet format
convert_nc_to_parquet("/content/drive/MyDrive/2_years_selected_variable/Transparency/cmems_obs-oc_atl_bgc-transp_my_l3-multi-1km_P1D_1.nc", "transparency1.parquet")
convert_nc_to_parquet("/content/drive/MyDrive/2_years_selected_variable/Transparency/cmems_obs-oc_atl_bgc-transp_my_l3-multi-1km_P1D_2.nc", "transparency2.parquet")
convert_nc_to_parquet("/content/drive/MyDrive/2_years_selected_variable/Transparency/cmems_obs-oc_atl_bgc-transp_my_l3-multi-1km_P1D_3.nc", "transparency3.parquet")
```

```python
parquet_files = [
    "transparency1.parquet",
    "transparency2.parquet",
    "transparency3.parquet",
]

tables = [pq.read_table(file) for file in parquet_files]
merged_table = pa.concat_tables(tables)

merged_parquet_path = "transparency_merged_ocean_data.parquet"
pq.write_table(merged_table, merged_parquet_path)

print(f"Merged dataset saved as {merged_parquet_path}")
```

```python
#Load Parquet files into DataFrames
df_transparency = pd.read_parquet("/content/drive/MyDrive/2_years_selected_variable/Transparency/transparency_merged_ocean_data.parquet")
```

## Plankton file converted to parquet and merged

```python
#Define function to convert NetCDF to Parquet
def convert_nc_to_parquet(nc_file, parquet_file):
    ds = xr.open_dataset(nc_file)
    df = ds.to_dataframe().reset_index()
    table = pa.Table.from_pandas(df)
    pq.write_table(table, parquet_file)
    print(f"Converted {nc_file} to {parquet_file}")

#Convert each dataset to Parquet format
convert_nc_to_parquet("/content/drive/MyDrive/2_years_selected_variable/Plankton/cmems_obs-oc_atl_bgc-plankton_my_l3-multi-1km_P1D_1.nc", "plankton1.parquet")
convert_nc_to_parquet("/content/drive/MyDrive/2_years_selected_variable/Plankton/cmems_obs-oc_atl_bgc-plankton_my_l3-multi-1km_P1D_2.nc", "plankton2.parquet")
convert_nc_to_parquet("/content/drive/MyDrive/2_years_selected_variable/Plankton/cmems_obs-oc_atl_bgc-plankton_my_l3-multi-1km_P1D_3.nc", "plankton3.parquet")
```

```python
parquet_files = [
    "plankton1.parquet",
    "plankton2.parquet",
    "plankton3.parquet",
]

# Read the schema of the first table
schema = pq.read_schema(parquet_files[0])

# Read all tables, ensuring they conform to the first schema
tables = [pq.read_table(file, schema=schema) for file in parquet_files]

merged_table = pa.concat_tables(tables)

merged_parquet_path = "plankton_merged_ocean_data.parquet"
pq.write_table(merged_table, merged_parquet_path)

print(f"Merged dataset saved as {merged_parquet_path}")
```

```python
#Load Parquet files into DataFrames
df_plankton = pd.read_parquet("plankton_merged_ocean_data.parquet")
```
